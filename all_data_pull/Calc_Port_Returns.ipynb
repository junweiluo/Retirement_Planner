```python
#Read In First Equity Dataframe Prices
first_equity_rtns = pd.read_csv('Stock ETF data pull.csv')
first_equity_rtns.head()
```

```python
#Read In Fixed Income Dataframe Prices
fixed_income_rtns = pd.read_csv('final_fixed_income.csv')
fixed_income_rtns.head()
```

```python
#Setting Date Column = Date
fixed_income_clean = fixed_income_final.set_index('Date')
fixed_income_clean.head()
```

```python
#Resetting the Dataframe with the dropped column
new_fixed_income = fixed_income_clean.reset_index()
new_fixed_income
```

```python
#Dropping the Date Column in Fixed Income DataFrame 
new_fixed_income_1 = new_fixed_income.drop(columns = ['Date'])
new_fixed_income_1
```

```python
#Concat all DataFrames for Full Universe
rtns_universe = pd.concat([new_first_equity,second_equity_rtns,new_fixed_income_1],axis = 'columns', join = 'inner')
rtns_universe
```

```python
#Setting the Blended Optimized ETFs
blended_universe = rtns_universe[['SMH', 'XLU','QQQ','TLT']]
blended_universe
```

```python
#Percent Change for the Blended Optimized Universe
blended_returns = blended_universe.pct_change()
blended_returns
```

```python
#Blended Daily Mean Returns
blended_daily_rtns = blended_returns.mean()
blended_daily_rtns
```

```python
#Blended Std
blended_std = blended_returns.std()
blended_std
```

```python
#Creating the Blended Optimized Portfolio DataFrame
blended_port_df = pd.concat([blended_daily_rtns, blended_std],axis = 'columns', join = 'inner')
blended_port_df
```

```python
#Cleaning the Blended Optimized Portfolio DataFramce --> Renaming Columns 
blended_port_final_df = blended_port_df.rename(columns = {0:'Mean Daily Returns', 1: 'Standard Deviation'})
blended_port_final_df
```

```python
#Setting the Final DataFrame 
df1 = blended_port_df.rename(columns = {0:'Mean Daily Returns', 1: 'Standard Deviation'})
df2 = pd.DataFrame({'Tickers':['SMH', 'XLU', 'QQQ', 'TLT'], 
                    'Weights':[0.22876, 0.20055, 0.31059, 0.2601]})
df3 = df2.set_index('Tickers')
blended_opt_weights = pd.concat([df1,df3], axis = 'columns', join='inner').T
blended_opt_weights
```
